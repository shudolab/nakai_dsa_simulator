```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# 日本語フォントの設定
matplotlib.rcParams['font.family'] = 'IPAexGothic'


# b01で始まるファイルのwinRateListをグラフに描画する
for i in range(6):
    df = pd.read_csv(f"../1m_simple/b045_at{(4*i+1)*60}.csv",header=None)
    # j列目を指定
    j = 0 # 列のインデックス

    # グラフを描画
    plt.plot(df.index + 1, df.iloc[:, j], marker='.', label=f"at{(4*i+1)*60}")  # +1 は行のインデックスを1から始めるため

#収束スピードを計算する

# 凡例を表示
plt.legend()

plt.xlabel("The number of simulations")
plt.ylabel(f"Attacker average win rate")
plt.title(f"atごとの勝率の収束スピード")
plt.show()
```
