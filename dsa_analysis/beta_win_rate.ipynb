```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 日本語フォントの設定
matplotlib.rcParams['font.family'] = 'IPAexGothic'
```

```python
# bの値のリスト
b_values = ['b01', 'b015', 'b02', 'b025', 'b03', 'b035', 'b04', 'b045']
for i in range(6):
    beta_distribution = []
    values = []
    # j列目を指定
    j = 0 # 列のインデックス
    for b in b_values:
        df = pd.read_csv(f"../1m_simple/{b}_at{(4*i+1)*60}.csv",header=None)
        beta_value = float(b[1]+"." +b[2:])
        beta_distribution.append(1/beta_value**2)
        values.append(df.iloc[-1,0])

    # グラフを描画
    plt.plot(beta_distribution, values, label=f"at{(4*i+1)*60}")

# 凡例を表示
plt.legend()

plt.xlabel("Distribution of attacker's block generation time")
plt.ylabel(f"Attacker average win rate")
plt.title(f"atごとの攻撃者のブロック生成時間の分布に対する攻撃者の勝率の変化")
plt.show()
```
